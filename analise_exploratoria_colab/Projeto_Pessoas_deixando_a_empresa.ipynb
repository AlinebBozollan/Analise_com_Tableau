4. Crie três maneiras diferentes de mostrar esses dados e liste quais são algumas vantagens e insuficiências de cada uma.

 - Criação de Gráficos e Visualizações.

Três Maneiras Alternativas de Visualização:

(a). Tabela de Texto:

![image.jpg](https://github.com/AlinebBozollan/Analise_com_Tableau/blob/main/Imagens/07.jpg?raw=true)

* Vantagens:

Clareza: As tabelas de texto são extremamente claras em exibir informações em formato tabular, tornando fácil a leitura dos dados.
Precisão: Elas permitem que os números sejam apresentados com alta precisão, evitando qualquer distorção visual.
Facilidade de Comparação: Os valores podem ser comparados de maneira direta, especialmente se houver uma ordem lógica nas linhas ou colunas.
Texto Explicativo: Pode incluir explicações detalhadas para cada célula, se necessário.

* Insuficiências:

Falta de Visualização: As tabelas de texto não têm o apelo visual de gráficos, tornando mais difícil identificar padrões ou tendências.
Dificuldade em Destacar: Pode ser difícil destacar rapidamente os valores mais altos ou mais baixos, especialmente em tabelas extensas.
Cansaço Visual: Em tabelas muito grandes, pode ocorrer fadiga visual devido à densidade das informações.

(b). Mapas de Árvore:

![image.jpg](https://github.com/AlinebBozollan/Analise_com_Tableau/blob/main/Imagens/05.jpg?raw=true)

* Vantagens:

Hierarquia Visual: Os mapas de árvore podem mostrar relações hierárquicas de maneira eficaz, destacando a estrutura das categorias.
Exibição de Fluxo: São excelentes para mostrar como os valores são divididos em subcategorias e, assim, visualizar o fluxo de dados.
Destaque de Valores: Permitem destacar valores-chave, mostrando-os como ramos principais.

* Insuficiências:

Complexidade: Em estruturas muito complexas, os mapas de árvore podem se tornar confusos e difíceis de interpretar.
Limitação de Dados: Não são adequados para exibir grandes quantidades de dados ou quando a hierarquia é muito profunda.
Dificuldade de Comparação: Comparar ramos em diferentes partes da árvore pode ser complicado, especialmente quando as ramificações são numerosas.

(c). Círculos Lado a Lado:

![image.jpg](https://github.com/AlinebBozollan/Analise_com_Tableau/blob/main/Imagens/06.jpg?raw=true)

* Vantagens:

Comparação Visual: Os círculos lado a lado permitem a comparação direta de tamanhos relativos, facilitando a identificação de diferenças.
Partição Proporcional: São eficazes em mostrar proporções e como as partes se somam a um todo.
Visualmente Atraente: São atraentes visualmente e podem ser usados para criar um impacto visual.

* Insuficiências:

Limitação de Dados: Se houver muitas categorias, os círculos podem se tornar pequenos e difíceis de distinguir.
Dificuldade em Mostrar Números Exatos: Embora permitam a comparação visual, não é possível obter números precisos a partir deles.
Não Adequados para Dados Temporais: Não são ideais para mostrar tendências ao longo do tempo ou mudanças sequenciais.

Em última análise, a escolha do melhor tipo de gráfico depende do objetivo da visualização, do conjunto de dados e do público-alvo. Cada tipo de gráfico tem suas próprias vantagens e limitações, e a seleção deve ser feita com base na eficácia em comunicar as informações desejadas de maneira clara e compreensível.
